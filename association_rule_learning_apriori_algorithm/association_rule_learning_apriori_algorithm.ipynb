### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import association_rules, apriori
from mlxtend.preprocessing import TransactionEncoder
```

### Load the data

```python
%cd /content/drive/MyDrive/python_for_business_and_finance/association_rule_learning_apriori_algorithm
df = pd.read_csv('groceries.csv')
df.head()
```

### Create transaction list

```python
transactions = []

for i in range(len(df)): # iterate over DataFrame and append items from each transaction to transactions list
  transaction = []
  for j in range(len(df.columns)):
    transaction.append(str(df.values[i, j]))
  transactions.append(transaction)
```

```python
transactions[0]
```

### Encode transactions

```python
enc = TransactionEncoder()
transactions_encoded = enc.fit_transform(transactions)

# Creata DataFrame from encoded transactions
df = pd.DataFrame(data = transactions_encoded, columns = enc.columns_)
df.head()
```

```python
# Remove NaN column
df.drop(columns = 'nan', inplace = True)
```

### Apriori

```python
model = apriori(df,
                min_support = 0.03,
                use_colnames = True)

results = association_rules(model,
                            metric = 'confidence',
                            min_threshold = 0.3)

# Sort results by lift
results.sort_values(by = 'lift', ascending = False)
```

### Visualization

```python
plt.scatter(results['support'], results['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
for i, label in enumerate(results.index):
  plt.annotate(label, (results['support'][i], results['confidence'][i]))
plt.show()
```
